```python
import torch
```

```python
batch_size = 4
window_size = 3
neg_count = 5
dim = 10
center_idx = 3
```

```python
S = torch.arange(batch_size * (2 * window_size + 1) * dim).reshape(batch_size, 
                                                                   2 * window_size + 1, 
                                                                   dim)
S
```

```python
center_words = S[:, center_idx, :].unsqueeze(1)
center_words, center_words.shape
```

```python
context_mask = torch.ones(S.shape[1])
context_mask = context_mask.type(torch.ByteTensor)
context_mask[center_idx] = 0
context_mask, context_mask.shape
```

```python
pos_contexts = S[:, context_mask, :]
pos_contexts, pos_contexts.shape
```

```python
pos_dots = torch.bmm(center_words, pos_contexts.transpose(1, 2))
pos_dots, pos_dots.shape
```
