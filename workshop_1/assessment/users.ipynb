```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_json('users.json', orient='records')
df.head()
```

```python
stats_df = pd.DataFrame()
```

```python
stats_df['tasks'] = df.groupby('tg_username')['task_number'].apply(list)
stats_df['count'] = df.groupby('tg_username')['task_number'].count()
stats_df = stats_df.sort_values('count', ascending=False)
stats_df
```
